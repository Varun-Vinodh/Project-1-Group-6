```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing csv files 

toronto_2015 = "Resources/toronto_climate_2015.csv"
toronto_2016 = "Resources/toronto_climate_2016.csv"
toronto_2017 = "Resources/toronto_climate_2017.csv"
toronto_2018 = "Resources/toronto_climate_2018.csv"
toronto_2019 = "Resources/toronto_climate_2019.csv"
```

```python
vancouver_2015 = "Resources/vancouver_climate_2015.csv"
vancouver_2016 = "Resources/vancouver_climate_2016.csv"
vancouver_2017 = "Resources/vancouver_climate_2017.csv"
vancouver_2018 = "Resources/vancouver_climate_2018.csv"
vancouver_2019 = "Resources/vancouver_climate_2019.csv"
```

```python
montreal_2015 = "Resources/montreal_climate_2015.csv"
montreal_2016 = "Resources/montreal_climate_2016.csv"
montreal_2017 = "Resources/montreal_climate_2017.csv"
montreal_2018 = "Resources/montreal_climate_2018.csv"
montreal_2019 = "Resources/montreal_climate_2019.csv"
```

```python
halifax_2015 = "Resources/halifax_climate_2015.csv"
halifax_2016 = "Resources/halifax_climate_2016.csv"
halifax_2017 = "Resources/halifax_climate_2017.csv"
halifax_2018 = "Resources/halifax_climate_2018.csv"
halifax_2019 = "Resources/halifax_climate_2019.csv"
```

```python
# reading csv files

toronto_15_df = pd.read_csv(toronto_2015)
toronto_16_df = pd.read_csv(toronto_2016)
toronto_17_df = pd.read_csv(toronto_2017)
toronto_18_df = pd.read_csv(toronto_2018)
toronto_19_df = pd.read_csv(toronto_2019)
```

```python
vancouver_15_df = pd.read_csv(vancouver_2015)
vancouver_16_df = pd.read_csv(vancouver_2016)
vancouver_17_df = pd.read_csv(vancouver_2017)
vancouver_18_df = pd.read_csv(vancouver_2018)
vancouver_19_df = pd.read_csv(vancouver_2019)
```

```python
montreal_15_df = pd.read_csv(montreal_2015)
montreal_16_df = pd.read_csv(montreal_2016)
montreal_17_df = pd.read_csv(montreal_2017)
montreal_18_df = pd.read_csv(montreal_2018)
montreal_19_df = pd.read_csv(montreal_2019)
```

```python
halifax_15_df = pd.read_csv(halifax_2015)
halifax_16_df = pd.read_csv(halifax_2016)
halifax_17_df = pd.read_csv(halifax_2017)
halifax_18_df = pd.read_csv(halifax_2018)
halifax_19_df = pd.read_csv(halifax_2019)
```

```python
# Concatenating data into a single dataframe
```

```python
frames = [toronto_15_df, toronto_16_df, toronto_17_df, toronto_18_df, toronto_19_df, \
          vancouver_15_df, vancouver_16_df, vancouver_17_df, vancouver_18_df, vancouver_19_df, \
          montreal_15_df, montreal_16_df, montreal_17_df, montreal_18_df, montreal_19_df, \
          halifax_15_df, halifax_16_df, halifax_17_df, halifax_18_df, halifax_19_df]   
```

```python
weather_concat = pd.concat(frames)
```

```python
weather_concat
```

```python
# Selecting columns

weather_concat.columns
```

```python
weather_cleaned = weather_concat[["Station Name","Month","Mean Temp (°C)","Total Precip (mm)"]]
```

```python
weather_cleaned
```

```python
# grouping data by "City"and "Month" and finding the average temperature and precipitation

weather_avg = weather_cleaned.groupby(by=["Station Name","Month"]).mean()
```

```python
weather_avg_df = pd.DataFrame(weather_avg)
weather_avg_df
```

```python
# resetting index
weather_avg_df = weather_avg_df.reset_index()
```

```python
# Renaming columns
weather_city = weather_avg_df.rename(columns={"Station Name":"City","Mean Temp (°C)":"Avg Temp (°C)","Total Precip (mm)":"Avg Precip (mm)"})
```

```python
weather_city
```

```python
# Adjusting city names
weather_city = weather_city.replace({"HALIFAX STANFIELD INT'L A":"Halifax" , "MONTREAL/PIERRE ELLIOTT TRUDEAU INTL":"Montreal" , "TORONTO CITY":"Toronto" , "VANCOUVER INTL A":"Vancouver"})
```

```python
weather_city
```

```python
# Adjusting data for quarterly frequency

weather_city["Quarter"] = weather_city["Month"].replace({1:"Q1",2:"Q1",3:"Q1", \
                                         4:"Q2",5:"Q2",6:"Q2", \
                                         7:"Q3",8:"Q3",9:"Q3", \
                                         10:"Q4",11:"Q4",12:"Q4"})
```

```python
weather_city_df = weather_city[["City","Quarter","Avg Temp (°C)","Avg Precip (mm)"]]
```

```python
weather_city_df
```

```python
# finding the average temperature and precipitation per quarter

weather_city_quarterly = weather_city_df.groupby(by=["City","Quarter"]).mean()
```

```python
weather_city_quarterly
```

```python
# resetting index and rounding values
weather_city_quarterly = weather_city_quarterly.reset_index()
```

```python
weather_city_quarterly.round(decimals=2)
```

```python
weather_city_quarterly
```

```python
# assigning values to lists

cities = ["Halifax","Montreal","Toronto","Vancouver"]
```

```python
temperature = []
precipitation = [] 
```

```python
for city in cities:
    temp = weather_city_quarterly.loc[weather_city_quarterly["City"] == city, ["City","Quarter","Avg Temp (°C)"]]
    precip = weather_city_quarterly.loc[weather_city_quarterly["City"] == city, ["City","Quarter","Avg Precip (mm)"]]
    temperature.append(temp)
    precipitation.append(precip)
```

```python
temperature
```

```python
precipitation
```

```python
# plotting average temperture changes per city

plt.plot(temperature[0]["Quarter"], temperature[0]["Avg Temp (°C)"])
plt.plot(temperature[1]["Quarter"], temperature[1]["Avg Temp (°C)"])
plt.plot(temperature[2]["Quarter"], temperature[2]["Avg Temp (°C)"])
plt.plot(temperature[3]["Quarter"], temperature[3]["Avg Temp (°C)"])

plt.title("Average temperature per city (2015 to 2019) - Quarterly")
plt.xticks(weather_city_quarterly["Quarter"], ["Jan-Feb-Mar","Apr-May-Jun","Jul-Aug-Sep","Oct-Nov-Dec"], rotation=15)
# plt.ylim(155, 205, 10)
plt.ylabel("Temperature (C)")
plt.legend(cities)
plt.grid()

plt.savefig("output/temperature_change.png", dpi=100)
```

```python
# plotting average precipitation changes per city

plt.plot(precipitation[0]["Quarter"], precipitation[0]["Avg Precip (mm)"])
plt.plot(precipitation[1]["Quarter"], precipitation[1]["Avg Precip (mm)"])
plt.plot(precipitation[2]["Quarter"], precipitation[2]["Avg Precip (mm)"])
plt.plot(precipitation[3]["Quarter"], precipitation[3]["Avg Precip (mm)"])

plt.title("Average precipitation per city (2015 to 2019) - Quarterly")
plt.xticks(weather_city_quarterly["Quarter"], ["Jan-Feb-Mar","Apr-May-Jun","Jul-Aug-Sep","Oct-Nov-Dec"], rotation=15)
plt.ylabel("Precipitation (mm)")
plt.legend(cities)
plt.grid()

plt.savefig("output/precipitation_change.png", dpi=100)
```

```python
# chart for Halifax 

fig, ax1 = plt.subplots()

ax1.plot(temperature[0]["Quarter"], temperature[0]["Avg Temp (°C)"], color="blue")
ax1.set_ylabel("Temperature (C)")
plt.ylim(-10, 25, 10)


ax2 = ax1.twinx() 

ax2.bar(precipitation[0]["Quarter"], precipitation[0]["Avg Precip (mm)"], color="blue", alpha=0.5)
ax2.set_ylabel("Precipitation (mm)")
plt.ylim(0, 10 , 2)

plt.title("Avg temp. and precip. - Halifax (2015 to 2019) - Quarterly")
plt.xticks(weather_city_quarterly["Quarter"], ["Jan-Feb-Mar","Apr-May-Jun","Jul-Aug-Sep","Oct-Nov-Dec"], rotation=15)

plt.grid()

fig.tight_layout() 

plt.savefig("output/halifax_weather.png", dpi=100)
```
