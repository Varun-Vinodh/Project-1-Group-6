# Canada Hotel Rates - EDA

Data Source: http://www.mtc-currentperformance.com/Hotel.aspx


```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing csv files 
tor_hotel = "Resources/Toronto_HotelIndustry_2015to2019.csv"
mon_hotel = "Resources/Montreal_HotelIndustry_2015to2019.csv"
van_hotel = "Resources/Vancouver_HotelIndustry_2015to2019.csv"
hal_hotel = "Resources/Halifax_HotelIndustry_2015to2019.csv"
```

```python
# reading csv files
tor_hotel_df = pd.read_csv(tor_hotel)
mon_hotel_df = pd.read_csv(mon_hotel)
van_hotel_df = pd.read_csv(van_hotel)
hal_hotel_df = pd.read_csv(hal_hotel)
```

```python
#concatenating into a singe data frame
concat = [tor_hotel_df,\
         mon_hotel_df,\
         van_hotel_df,\
         hal_hotel_df]
```

```python
hotel_concat = pd.concat(concat)
```

```python
# print concatenated data frame
hotel_concat
```

```python
#Clean the data frame and rename the column names 
hotel_clean = hotel_concat[['Geo', 'Year', ' Quarter', ' Occupancy Rate', ' Daily Rate']]
hotel_clean.rename(columns={'Geo': 'City','Year':'Year', ' Quarter': 'Quarter', ' Occupancy Rate': 'Average Occupancy Rate (%)',
                                             ' Daily Rate': 'Average Daily Rate ($)'}, inplace=True)
```

```python
#remove $ and % symbols and convert the values to float
hotel_clean['Average Daily Rate ($)'] = hotel_clean['Average Daily Rate ($)'].str.replace('$', '').astype(float)
hotel_clean['Average Occupancy Rate (%)'] = hotel_clean['Average Occupancy Rate (%)'].replace({'%':''}, regex=True).astype(float)
```

```python
#Cleaned data frame
hotel_clean_df = hotel_clean[['City','Quarter','Average Daily Rate ($)','Average Occupancy Rate (%)']]
```

```python
#find the mean value to get Average Daily Rate and Average Occupancy Rate
avg_hotelrate = hotel_clean_df.groupby(by=["City","Quarter"]).mean()
```

```python
#print the average hotel rates
avg_hotelrate
```

```python
# reseting index
dataframe = avg_hotelrate.reset_index()
dataframe
```

```python
#plot the line graph for the Average Daily Rate 
cities = ["Halifax", "Montreal", "Toronto", "Vancouver"]
city_avg_hotelrate = []
for city in cities:
    price_data = dataframe.loc[dataframe["City"] == city, ["City","Quarter","Average Daily Rate ($)","Average Occupancy Rate (%)"]]
    city_avg_hotelrate.append(price_data)

plt.plot(city_avg_hotelrate[0]["Quarter"], city_avg_hotelrate[0]["Average Daily Rate ($)"])
plt.plot(city_avg_hotelrate[1]["Quarter"], city_avg_hotelrate[1]["Average Daily Rate ($)"])
plt.plot(city_avg_hotelrate[2]["Quarter"], city_avg_hotelrate[2]["Average Daily Rate ($)"])
plt.plot(city_avg_hotelrate[3]["Quarter"], city_avg_hotelrate[3]["Average Daily Rate ($)"])

plt.title("Average Hotel rate per city (2015 to 2019) - Quarterly")
plt.xticks(dataframe["Quarter"], ["Jan-Feb-Mar","Apr-May-Jun","Jul-Aug-Sep","Oct-Nov-Dec"])
plt.ylabel("Average Daily Hotel Rates ($)")
plt.legend(cities)
plt.grid()

#save the graph
plt.savefig("output/average_hotel_rate.png", dpi=100)
```
