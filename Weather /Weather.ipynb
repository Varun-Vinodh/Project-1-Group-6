```python
#Import Dependencies
import pandas as pd
```

```python
halifax_2015 = "Resources/2015/halifax_climate_2015.csv"
montreal_2015 = "Resources/2015/montreal_climate_2015.csv" 
toronto_2015 = "Resources/2015/toronto_climate_2015.csv"
vancouver_2015 = "Resources/2015/vancouver_climate_2015.csv"
```

```python
halifax_2015_df = pd.read_csv(halifax_2015)
montreal_2015_df = pd.read_csv(montreal_2015)
toronto_2015_df = pd.read_csv(toronto_2015)
vancouver_2015_df = pd.read_csv(vancouver_2015)
halifax_2015_df.head()
```

```python
halifax_2015_df.columns
```

```python
#cleaning up the datasets 
halifax_2015_df = halifax_2015_df[["Date/Time",'Station Name','Mean Temp (°C)','Total Precip (mm)',]]
toronto_2015_df = toronto_2015_df[["Date/Time",'Station Name','Mean Temp (°C)','Total Precip (mm)',]]
montreal_2015_df = montreal_2015_df[["Date/Time",'Station Name','Mean Temp (°C)','Total Precip (mm)',]]
vancouver_2015_df = vancouver_2015_df[["Date/Time",'Station Name','Mean Temp (°C)','Total Precip (mm)',]]
```

```python
halifax_2015_df.head()
```

```python
#Halifax 2015
#finding Halifax's mean temp and temp total precip for the four quarters of the year 

# 1st quarter Jan-Feb-Mar
first_quarter_halifax_2015 = halifax_2015_df.loc[(halifax_2015_df["Date/Time"] <= "2015-03-31"),:]
avg_halifax_first_quarter = first_quarter_halifax_2015.groupby("Station Name").mean()

#2nd quarter Apr-May_Jun
second_quarter_halifax_2015 = halifax_2015_df.loc[((halifax_2015_df["Date/Time"] >= "2015-04-01") 
                                                   & (halifax_2015_df["Date/Time"] <= "2015-06-30")),:]
avg_halifax_second_quarter = second_quarter_halifax_2015.groupby("Station Name").mean() 

#3rd quarter Jul-Aug-Sep
third_quarter_halifax_2015 = halifax_2015_df.loc[((halifax_2015_df["Date/Time"] >= "2015-07-01") 
                                                   & (halifax_2015_df["Date/Time"] <= "2015-09-30")),:]
avg_halifax_third_quarter = third_quarter_halifax_2015.groupby("Station Name").mean() 

#4nd quarter Oct-Nov-Dec
fourth_quarter_halifax_2015 = halifax_2015_df.loc[((halifax_2015_df["Date/Time"] >= "2015-10-01") 
                                                   & (halifax_2015_df["Date/Time"] <= "2015-12-31")),:]
avg_halifax_fourth_quarter = fourth_quarter_halifax_2015.groupby("Station Name").mean() 

#creating a dataframe with the data

halifax_2015_avg_df = pd.concat([avg_halifax_first_quarter, avg_halifax_second_quarter,
                                   avg_halifax_third_quarter,avg_halifax_fourth_quarter], 
                                keys =("Jan-Feb-Mar","Apr-May_Jun","Jul-Aug-Sep","Oct-Nov-Dec"))

halifax_2015_avg_df
```
