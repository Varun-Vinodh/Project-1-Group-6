```python
halifax_15_df = pd.read_csv(halifax_2015)
halifax_16_df = pd.read_csv(halifax_2016)
halifax_17_df = pd.read_csv(halifax_2017)
halifax_18_df = pd.read_csv(halifax_2018)
halifax_19_df = pd.read_csv(halifax_2019)
```

```python
frames = [toronto_15_df, toronto_16_df, toronto_17_df, toronto_18_df, toronto_19_df, \
          vancouver_15_df, vancouver_16_df, vancouver_17_df, vancouver_18_df, vancouver_19_df, \
          montreal_15_df, montreal_16_df, montreal_17_df, montreal_18_df, montreal_19_df, \
          halifax_15_df, halifax_16_df, halifax_17_df, halifax_18_df, halifax_19_df]   
```

```python
weather_concat = pd.concat(frames)
```

```python
weather_concat
```

```python
weather_concat.columns
```

```python
weather_cleaned = weather_concat[["Station Name","Month","Mean Temp (Â°C)","Total Precip (mm)"]]
```

```python
weather_cleaned
```

```python
weather_avg = weather_cleaned.groupby(by=["Station Name","Month"]).mean()
```

```python
weather_avg_df = pd.DataFrame(weather_avg)
weather_avg_df
```

```python
weather_avg_df = weather_avg_df.reset_index()
```

```python
weather_city = weather_avg_df.rename(columns={"Station Name":"City","Mean Temp (Â°C)":"Avg Temp (Â°C)","Total Precip (mm)":"Avg Precip (mm)"})
```

```python
weather_city
```

```python
weather_city = weather_city.replace({"HALIFAX STANFIELD INT'L A":"Halifax" , "MONTREAL/PIERRE ELLIOTT TRUDEAU INTL":"Montreal" , "TORONTO CITY":"Toronto" , "VANCOUVER INTL A":"Vancouver"})
```

```python
weather_city
```

```python
# weather_city["Month"]
```

```python
weather_city["Quarter"] = weather_city["Month"].replace({1:"Q1",2:"Q1",3:"Q1", \
                                         4:"Q2",5:"Q2",6:"Q2", \
                                         7:"Q3",8:"Q3",9:"Q3", \
                                         10:"Q4",11:"Q4",12:"Q4"})
```

```python
weather_city_df = weather_city[["City","Quarter","Mean Temp (Â°C)","Total Precip (mm)"]]
```

```python
weather_city_df
```

```python
weather_city_quarterly = weather_city_df.groupby(by=["City","Quarter"]).mean()
```

```python
weather_city_quarterly
```

```python
weather_city_quarterly = weather_city_quarterly.reset_index()
```

```python
weather_city_quarterly.round(decimals=2)
```

```python
weather_city_quarterly
```

```python
cities = ["Halifax","Montreal","Toronto","Vancouver"]
```

```python
temperature = []
precipitation = [] 
```

```python
for city in cities:
    temp = weather_city_quarterly.loc[weather_city_quarterly["City"] == city, ["City","Quarter","Mean Temp (Â°C)"]]
    precip = weather_city_quarterly.loc[weather_city_quarterly["City"] == city, ["City","Quarter","Total Precip (mm)"]]
    temperature.append(temp)
    precipitation.append(precip)
```

```python
temperature
```

```python
plt.plot(temperature[0]["Quarter"], temperature[0]["Mean Temp (Â°C)"])
plt.plot(temperature[1]["Quarter"], temperature[1]["Mean Temp (Â°C)"])
plt.plot(temperature[2]["Quarter"], temperature[2]["Mean Temp (Â°C)"])
plt.plot(temperature[3]["Quarter"], temperature[3]["Mean Temp (Â°C)"])

plt.title("Average temperature per city (2015 to 2019) - Quarterly")
plt.xticks(weather_city_quarterly["Quarter"], ["Jan-Feb-Mar","Apr-May-Jun","Jul-Aug-Sep","Oct-Nov-Dec"], rotation=15)
# plt.ylim(155, 205, 10)
plt.ylabel("Temperature (C)")
plt.legend(cities)
plt.grid()

plt.show()
```

```python
precipitation
```

```python
plt.bar(precipitation[0]["Quarter"], precipitation[0]["Total Precip (mm)"])

plt.title("Average precipitation per city (2015 to 2019) - Quarterly")
plt.xticks(weather_city_quarterly["Quarter"], ["Jan-Feb-Mar","Apr-May-Jun","Jul-Aug-Sep","Oct-Nov-Dec"], rotation=15)
# plt.ylim(155, 205, 10)
plt.ylabel("Precipitation (mm)")
plt.legend(cities)
plt.grid()

plt.show()
```

```python
plt.bar(precipitation[1]["Quarter"], precipitation[1]["Total Precip (mm)"], color="orange")

plt.title("Average precipitation for Montreal (2015 to 2019) - Quarterly")
plt.xticks(weather_city_quarterly["Quarter"], ["Jan-Feb-Mar","Apr-May-Jun","Jul-Aug-Sep","Oct-Nov-Dec"], rotation=15)
# plt.ylim(155, 205, 10)
plt.ylabel("Precipitation (mm)")
# plt.legend(cities[1])
plt.grid()

plt.show()
```
